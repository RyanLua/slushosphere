## Data Table
Table with recorded data from the Arduino and sensors. Converts the epoch timestamp to a human-readable date format.

```python
import pandas

df = pandas.read_csv("../data/DATA.CSV")

df = df.rename(columns={"Timestamp (Epoch)": "Timestamp"})
df["Timestamp"] = pandas.to_datetime(df["Timestamp"], unit="s")
df.head()
```

## Sensor Data Graph
The following plots show the trends for each sensor value over time.

```python
import matplotlib.pyplot as plt

columns = [
    "Temperature (°C)",
    "Humidity (%)",
    "Pressure (kPa)",
    "Illuminance (lx)",
    "Altitude (m)",
    "Speed (km/h)",
    "Satellites",
]

fig, axs = plt.subplots(len(columns), 1, figsize=(12, 18), sharex=True)
for i, col in enumerate(columns):
    axs[i].plot(df["Timestamp"], df[col], label=col)
    axs[i].set_ylabel(col)

axs[-1].set_xlabel("Timestamp")
plt.tight_layout()
plt.show()
```
